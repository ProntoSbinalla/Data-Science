## Multiple Linear Regression and Adjusted R-Squared

Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set_theme()
```

Importing Data

```python
data = pd.read_csv('MultipleLinearRegression.csv')
data.describe()
```

Creating Multiple Regression

```python
y = data['GPA']
x1 = data[['SAT', 'Rand 1,2,3']]
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
rawData = pd.read_csv("DummyVariables.csv")
rawData
```

```python
data = rawData.copy()
data['Attendance'] = data['Attendance'].map({'Yes': 1, 'No':0})
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

### Making Predictions based on Regression we Created

```python
new_data = pd.DataFrame({'const': 1, 'SAT': [1700, 1670], 'Attendance': [0, 1]})
new_data = new_data[['const', 'SAT', 'Attendance']]
new_data.rename(index={0:'Bob', 1:'Alice'})
```

```python
predictions = results.predict(new_data)
predictions
```

```python
predictionsdf = pd.DataFrame({'Predictions': predictions})
joined = new_data.join(predictionsdf)
joined.rename(index={0:'Bob', 1:'Alice'})
```
